# EDA

### How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
- Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
# Create filte that shows a value greater than '0' in either column
filter_dg = (movie_df['budget'] > 0) | (movie_df['revenue'] > 0)
```

```python
# Use value_counts to see how many fall into this category
filter_dg.value_counts()
```

```python
# Create the reciprocal (set a variable where == 0 is true for both)
zero_financial = (movie_df['budget'] == 0) & (movie_df['revenue'] == 0)
```

```python
# Check against value counts
zero_financial.value_counts()
```

#### We can see that the opposite is true for our "&" filter. This is two ways to show that: 1881 movies have values greater than '0' in either "revenue", "budget", or both

### How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Again, a simple value_countscan solve this question
movie_df['certification'].value_counts()
```

```python
# Create a rating df with only "G, PG, PG-13, R" ratings
rating_df = movie_df.copy() # As not to further alter the original


rating_df['certification'] = rating_df['certification'].replace({'-':'drop',
                                                                'NC-17':'drop',
                                                                'NR':'drop',
                                                                'Unrated':'drop'})
# Filter out rows with 'drop' in the certification column
rating_df = rating_df[rating_df.certification != 'drop']
rating_df['certification'].value_counts()
```

#### We can see that:
- 453 movies are rated R
- 181 movies are rated PG-13
- 64 movies are rated PG
- 24 movies are rated G

```python
# Visual representation
rating_df["certification"].hist();
plt.ylabel("Number of Moves")
plt.xlabel("Certification/Rating");
```

### What is the average revenue per certification category?

```python
rating_df.groupby('certification')['revenue'].mean().round(0)
```

### Average Revenue by Certification
- G:     72,185,327
- PG:    61,435,347
- PG-13: 71,842,275
- R:     16,786,907

```python
# Visual representation of revenue distribution (in scientific notation)
sns.barplot(data = rating_df, x = "certification", y = "revenue");
plt.ylabel("Revenue(in scientific notation)")
plt.xlabel("Certification/Rating)");
```

### What is the average budget per certification category?

```python
rating_df.groupby('certification')['budget'].mean().round(0)
```

### Average Budget by Certification
- G:     23,833,333
- PG:    24,597,656
- PG-13: 31,231,536
- R:     9,976,831

```python
# Visual representation of budget distribution (in scientific notation)
sns.boxplot(x = 'certification', y = 'budget', data = rating_df);
```

#### Average budget by certification is:
- R = 9,976,831
- PG-13 = 31,231,536
